```python
from commitgen.diff import AddRemExtractor, PerFileExtractor, is_nl
from commitgen.data import RawDataset, extract_commits, parse_commits
from commitgen.code import CodeChunkTokenizer, CodeLinesTokenizer
from commitgen.nlp import SennaTokenizer, TreebankTokenizer
from pygments.token import Comment, String, Whitespace, Text
```

```python
marker = "NEW_FILE"
```

```python
language = "javascript"
```

```python
add_rem_code_extractor = AddRemExtractor(marker=marker)
per_file_code_extractor = PerFileExtractor(marker=marker)
```

```python
code_chunk_tokenizer = CodeChunkTokenizer(language=language)
code_lines_tokenizer = CodeLinesTokenizer(language=language)
```

```python
senna_tokenizer = SennaTokenizer()
treebank_tokenizer = TreebankTokenizer()
```

```python
raw_dataset_a = RawDataset("/home/ubuntu/data/react_commits/")
```

```python
raw_dataset_b = RawDataset("/home/ubuntu/data/node_commits/")
```

```python
# extract_filters = [lambda c: len(c.diff_file.modified_files) + 
#                              len(c.diff_file.added_files) +
#                              len(c.diff_file.removed_files) == 1]
```

```python
parse_filters = [lambda pc: 1 <= len(pc.code_tokens) <= 100,
                 lambda pc: 1 <= len(pc.nl_tokens) <= 100]
```

```python
ignore_list= [Comment, String, Whitespace, Text]
```

```python
commits_a = extract_commits(raw_dataset_a,
                            per_file_code_extractor,
                            filters=[])
print len(commits_a)
```

```python
commits_b = extract_commits(raw_dataset_b,
                            per_file_code_extractor,
                            filters=[])
print len(commits_b)
```

```python
parsed_commits_a = parse_commits(commits_a, treebank_tokenizer,
                                 code_chunk_tokenizer,
                                 filters=[],
                                 ignore_types=ignore_list)
print len(parsed_commits_a)
```

```python
parsed_commits_b = parse_commits(commits_b, treebank_tokenizer,
                                 code_chunk_tokenizer,
                                 filters=[],
                                 ignore_types=ignore_list)
print len(parsed_commits_b)
```

```python
from collections import Counter
words_a = Counter()
code_tokens_a = Counter()
for parsed_commit in parsed_commits_a:
    words_a.update(parsed_commit.nl_tokens)
    code_tokens_a.update(parsed_commit.code_tokens)
```

```python
from collections import Counter
words_b = Counter()
code_tokens_b = Counter()
for parsed_commit in parsed_commits_b:
    words_b.update(parsed_commit.nl_tokens)
    code_tokens_b.update(parsed_commit.code_tokens)
```

```python
s_words_a = set(words_a.keys())
s_code_tokens_a = set(code_tokens_a.keys()) 
s_words_b = set(words_b.keys())
s_code_tokens_b = set(code_tokens_b.keys())
```

```python
unique_words = s_words_a & s_words_b
```
